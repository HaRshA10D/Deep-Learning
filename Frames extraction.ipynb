```python
import cv2
import os
import numpy as np
```

```python
x = 1
for filename in os.listdir('data/depthvideos'):
    video_file = cv2.VideoCapture('data/depthvideos/'+filename)
    tot_frames = int(video_file.get(cv2.CAP_PROP_FRAME_COUNT))
    print(tot_frames)
    frame_array = []
    folder = 'data/train/' + filename[3:5]+'/'+ str(x)
    if not os.path.exists(folder):
        os.makedirs(folder)
        x += 1
    while True:
        val,frame = video_file.read()
        if val == False:
            break
        #sobelx = cv2.Sobel(frame,cv2.CV_64F,1,1,ksize=3)
        frame_array.append(frame)
    frame_array = np.array(frame_array)
    if tot_frames > 32:
        frame_valid_no = np.random.choice(range(1,17),2,replace=False)
        frame_valid_no.sort()
        yb = 1
        for i in frame_valid_no:
            cv2.imwrite(folder+'/frame-%d.jpeg'%(yb),frame_array[i-1])
            yb += 1
    if tot_frames < 32:
        frame_valid_no = np.random.choice(range(1,tot_frames+1),32,replace=True)
        frame_valid_no.sort()
        yb = 1
        for i in frame_valid_no:
            cv2.imwrite(folder+'/frame-%d.jpeg'%(yb),frame_array[i-1])
            yb += 1
    if tot_frames == 32:
        for i in range(len(frame_array)):
            cv2.imwrite(folder+'/frame-%d.jpeg'%(i+1),frame_array[i])
```
